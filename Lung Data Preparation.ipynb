## Clean the data

```python
# https://www.kaggle.com/datasets/mysarahmadbhat/lung-cancer/

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
PATH = "Data\survey_lung_cancer.csv"
data = pd.read_csv(PATH)
data.head()

# 2 = YES, 1 = NO
# 2 = Male, 1 = Female
```

```python
data.info()
```

```python
data.isnull().sum()
```

## Transform GENDER and LUNG_CANCER into integers
1 = Yes, 0 = No

2 = Male, 1 = Female

```python
dictGender = {"M": 2, "F": 1}
dictYN = {"YES": 1, "NO": 0}

data.replace({"GENDER": dictGender}, inplace = True)
data.replace({"LUNG_CANCER": dictYN}, inplace = True)

data.head()
```

## Plot the relationships

```python
for i, feature in enumerate(data.columns):
    if feature != "LUNG_CANCER":
        plt.figure(i)
        sns.catplot(x = feature, y = "LUNG_CANCER", data = data, kind = "point", aspect = 2)
        plt.title(f"Lung Cancer Prevalence vs {feature}")
    
# plot one feature against lung cancer to get it right and then loop through the rest
```

```python
data.to_csv("Data\survey_lung_cancer_cleaned.csv", index = False)
```
